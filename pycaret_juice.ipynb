# **Looking all datasets of pycaret using index**

```python
from pycaret.datasets import get_data
get_data('index')
```

**Selected and Read Juice Dataset**

```python
data = get_data('juice')
```

```python
data.shape
```

```python
data['Purchase'].value_counts()
```

# **Transforming the Categorical Data into Numeric Using LabelEncoder**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data.Purchase = le.fit_transform(data.Purchase)
data.Store7 = le.fit_transform(data.Purchase)
data.head()
```

```python
data.isna().sum()
```

# **Training the data**

```python
from pycaret.classification import *
s = setup(data=data, target="Purchase", log_experiment=True,
          experiment_name='juicemlflow', log_plots=True, train_size=0.7,
          session_id=42)
```

```python
X_train = get_config('X_train')
X_test = get_config('X_test')
y_train = get_config('y_train')
y_test = get_config('y_test')

print(X_train.dtypes)  # Check column data types in X_train
print(X_test.dtypes)   # Check column data types in X_test
print(y_train.dtypes)  # Check data types in y_train
print(y_test.dtypes)   # Check data types in y_test
```

# **Comparing the models**

```python
# Compare models and select the best one automatically
best_model = compare_models()
```

# **Tuning the Model for Better Model Performance**

```python
tuned_model = tune_model(best_model)
print(tuned_model)
```

# **Finalizing the model**

```python
final_model = finalize_model(tuned_model)
print(final_model)
```

# **Model Evaluation to Check Model Performance**

```python
# Evaluate the model
evaluate_model(final_model)
```

# **Saving the Model**

```python
# Log model in MLflow and save it
mlflow_run = save_model(final_model, "juice_pycaret_model", model_only=True)
print("Model saved and logged in MLflow!")
```

# **Taking the best model for furthur training with it**

```python
model = create_model('lr')
```

```python
model.fit(X_train, y_train)
```

# **Predicting with model with test data**

```python
y_pred = model.predict(X_test)
```

# **Checking the Performance of our model using classification_report**

classification_report is one of the best evaluation emport for classification models as it consists of all the fundamentals metrics like precision, recall, f1-score.

```python
from sklearn.metrics import classification_report
report_dict = classification_report(y_test, y_pred, output_dict=True)
print(report_dict)
```
